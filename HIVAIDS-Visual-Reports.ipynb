```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r"C:\Users\user\Documents\Total Reported Cases per Year.csv")
display(dataset)
```

```python
Years = dataset["Demographic Year"]
ReportedCases = dataset["Total Reported Cases"]

rcy = plt.figure()
rcy.set_figwidth(12)
rcy.set_figheight(7)

plt.plot(Years, ReportedCases)
plt.title("Total Number of Reported Cases from 2015-2021", fontweight="bold")
plt.xlabel("Years")
plt.ylabel("Number of Reported Cases")
plt.grid(color = 'black', linestyle = '-', linewidth = 0.5)
plt.show()
```

```python
dataset = pd.read_csv(r"C:\Users\user\Documents\Total Reported Cases per Year.csv")
display(dataset.iloc[0:,[0,6,7,8,9,10]])
```

```python
Year = dataset["Demographic Year"]
A1 =dataset["Less than 15 y/o"]
A2 =dataset["15-24 y/o"]
A3 = dataset["25-34 y/o"]
A4 = dataset["35-49 y/o"]
A5 = dataset["50 y/o and above"]

rcy = plt.figure()
rcy.set_figwidth(12)
rcy.set_figheight(7)

barTry=0.11
X_axis=np.arange(len(Year))
plt.bar(X_axis -0.2,A1, color='r',width=barTry,label='Less than 15 y/o')
plt.bar(X_axis -0.1,A2, color='b',width=barTry,label='15-24 y/o')
plt.bar(X_axis,A3, color='purple',width=barTry,label='25-34 y/o')
plt.bar(X_axis +0.1,A4, color='g',width=barTry,label='35-49 y/o')
plt.bar(X_axis +0.2,A5, color='orange',width=barTry,label='50 y/o and above')
plt.xticks(X_axis,Year)

plt.title("Number of Reported Cases by their Ages from 2015-2021", fontweight="bold")
plt.xlabel("Years")
plt.ylabel("Number of Cases")

plt.grid(color = 'black', linewidth = 0.5)
plt.legend()
plt.show()
```

```python
dataset = pd.read_csv(r"C:\Users\user\Documents\Type of Transmission.csv")
display(dataset)
```

```python
Year = dataset["Demographic Year"]
B1 = dataset["Sexual Contact"]
B2 = dataset["Blood/Blood Products"]
B3 = dataset["Sharing of Needles"]
B4 = dataset["Needle Prick Injury"]
B5 = dataset["Mother to Child"]
B6 = dataset["No Data Available"]

rcy = plt.figure()
rcy.set_figwidth(12)
rcy.set_figheight(7)

Y_axis=np.arange(len(Year))
barTry=0.150
plt.bar(Y_axis -0.2 , B1, color='r',width = barTry, label='Sexual Contact')
plt.bar(Y_axis -0.1, B2, color='y',width = barTry, label='Blood/Blood Products')
plt.bar(Y_axis, B3, color='b',width = barTry, label='Sharing of Needles')
plt.bar(Y_axis +0.1, B4, color='orange',width = barTry, label='Needle Prick Injury')
plt.bar(Y_axis +0.2, B5, color='purple',width = barTry, label='Mother to Child')
plt.bar(Y_axis +0.3, B6, color='g',width = barTry, label='No Data Available')
plt.xticks(Y_axis,Year)
plt.title("Type of Transmission of HIV/AIDS from 2015-2021", fontweight="bold")
plt.xlabel("Years")
plt.ylabel("Number of Cases")
plt.grid(color = 'black', linestyle = '-', linewidth = 0.5)
plt.legend(bbox_to_anchor=(1.0, 1.0))
plt.show()
```
